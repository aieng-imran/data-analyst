### Subplots in columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Seaborn's built-in dataset
tips = sns.load_dataset("tips")

sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col = "smoker")
plt.show()
```

### Subplots in rows

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Seaborn's built-in dataset
tips = sns.load_dataset("tips")

sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", row = "smoker")
plt.show()
```

### Subplots in rows and columns

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading Seaborn's built-in dataset
tips = sns.load_dataset("tips")

sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", 
            col = "smoker", row = "time")
plt.show()
```

### Wrapping columns

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", 
            col = "day", col_wrap = 2)
plt.show()
```

### Ordering columns

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", col="day",
            col_wrap=2, col_order=["Thur", "Fri", "Sat", "Sun"])

plt.show()
```

## Customizing scatter plots

### Subgroups with point size

```python
import seaborn as sns
import matplotlib.pyplot as plt 

sns.relplot(x= "total_bill", y= "tip", data=tips, kind= "scatter"
            , size= "size")
plt.show()
```

### Point size and hue

```python
sns.relplot(x= "total_bill", y= "tip", data=tips, kind= "scatter"
            , size= "size", hue = "size")
plt.show()
```

### Subgroups with point style

```python
sns.relplot(x= "total_bill", y= "tip", data=tips, kind= "scatter"
            , size= "size", hue= "smoker", style= "smoker")
plt.show()
```

### Changing point transparency

```python
sns.relplot(x= "total_bill", y= "tip", data=tips, kind= "scatter"
            , alpha = 0.4)
plt.show()
```

## Introduction to line plots

```python
# https://so2.gsfc.nasa.gov/no2/pix/htmls/Lahore_csv.html
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lahore_pollution = pd.read_csv('./Data/OMNO2_Timeseries_Lahore.csv')
print(lahore_pollution.head(2))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x="Date(yyyy-mm-dd)", y="Mean",
            data=lahore_pollution[:10], kind="scatter")

plt.xticks(rotation= 45)
plt.show()
```

### Line plot

```python
sns.relplot(x="Date(yyyy-mm-dd)", y="Mean",
            data=lahore_pollution[:10], kind="line")

plt.xticks(rotation= 45) 
plt.show()
```
