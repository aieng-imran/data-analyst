## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

## Import Data

```python
# cs = pd.read_csv('./Data/computersales.csv')
# cs.head()

# We instead used seaborn's built-in dataset

crash_df = sns.load_dataset('car_crashes') 
crash_df.head()
```

## Distribution Plots

### Distribution Plot

```python
sns.distplot(crash_df['not_distracted']);
```

```python
# The line above is called "Kernel Density Estimation"

sns.distplot(crash_df['not_distracted'] , kde =  False, bins = 25);
```

### Joint Plot

### Regression Plot

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg');
```

### KDE

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'kde');
```

```python
### Hexogon

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex');
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol']);
```

### Pair Plots

```python
sns.pairplot(crash_df);
```

```python
# Testing with newer Seaborn built-in data

tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue = 'sex', palette = 'Blues');

# Note: hue is used with categorial data
```

## Rug Plot

```python
sns.rugplot(tips_df['tip']);

# Note: In rug plots,lines are more denser as comapred to Histogram
```

## Styling

```python
sns.set_style('ticks')

plt.figure(figsize = (8, 4));

sns.set_context('paper', font_scale = 1.5)

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg');
```

```python
sns.set_style('ticks');

plt.figure(figsize = (8, 4));

sns.set_context('poster', font_scale = 0.8);

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg');
```

```python
sns.set_style('ticks');

plt.figure(figsize = (8, 4));

sns.set_context('poster', font_scale = 0.8);

sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg');

sns.despine(left = True, bottom = True)
```

## Categorial Plots

### Bar Plots

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips_df,  estimator = np.median);

# Note: By default, the estimator of this plot is "mean"
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips_df,  estimator = np.cov);

# Note: By default, the estimator of this plot is "mean"
# Note: We can also pluggin our own function in place of estimator
```

### Count Plots

```python
sns.countplot(x = 'sex', data = tips_df);
```

## Box Plot

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex');

plt.legend(loc = 0);
```

## Violin Plot

```python
# It's a combination of box plot & KDE
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', split =  True);
```

## Strip Plot

```python
plt.figure(figsize = (8, 5))

sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True, hue = 'sex', dodge = True);
```

## Swarm Plot

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips_df);
```

## Palettes

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html

plt.figure(figsize = (8,6))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x = 'day', y ='total_bill', data = tips_df, hue = 'sex', palette = 'magma');

plt.legend(loc = 9);
```

## PairGrid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map(plt.scatter);
```

```python
iris_g.map_diag(plt.hist);
iris_g.map_offdiag(plt.scatter);
```

```python
iris_g.map_upper(plt.scatter);
iris_g.map_lower(sns.kdeplot);
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species',
                     x_vars = ['sepal_length', 'sepal_width'],
                     y_vars = ['petal_length', 'petal_width'])

iris_g.map(plt.scatter);
iris_g.add_legend();
```

## Facet Grid

```python
tips_df = sns.load_dataset('tips')

tips_fg = sns.FacetGrid(tips_df, col = 'time', hue = 'smoker', height = 4, aspect = 1.3,
                       col_order = ['Dinner', 'Lunch'], palette = 'Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor = 'w');

kws = dict(s = 50, linewidth = 0.5, edgecolor = 'w')
tips_fg = sns.FacetGrid(tips_df, col = 'sex', hue = 'smoker', height = 4, aspect = 1.3,
                       hue_order = ['Yes', 'No'],
                       hue_kws = dict(marker = ['^', 'v']))

tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws);
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col = 'subject', col_wrap = 5, height = 1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker = '.');
```

## Regression Plot

```python
plt.figure(figsize = (8,6))
sns.set_context('paper', font_scale = 1.4)
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips_df,
          scatter_kws = {'s' : 100, 'linewidth' : 0.5, 'edgecolor' : 'w'});
```

```python
sns.set_context('poster', font_scale = 0.5)

sns.lmplot(x = 'total_bill', y = 'tip', col = 'sex', row = 'time', data = tips_df);
```
