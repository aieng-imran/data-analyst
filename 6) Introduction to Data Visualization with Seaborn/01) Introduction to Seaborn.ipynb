```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Scatter Plot

```python
# Scatter Plot
height = [62, 64, 69, 75, 66, 68, 65, 71, 76, 73]
weight = [120, 136, 148, 175, 137, 165, 154, 172, 200, 187]

sns.scatterplot(x=height, y=weight)
plt.show()
```

### Count Plot

```python
# Count Plot

gender = ["Female", "Female", "Female", "Female", "Male", "Male", "Male",
"Male", "Male", "Male"]

sns.countplot(x=gender)
plt.show()
```

### Bar Chart

### Adding a third variable with hue

```python
# Using seaborn's built-in datasets

import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

### A scatter plot with hue

```python
sns.scatterplot(x= "total_bill", y= "tip", data=tips, hue= "smoker")
plt.show()
```

### Setting hue order

```python
sns.scatterplot(x= "total_bill", y= "tip", data=tips, hue= "smoker"
                , hue_order=["Yes", "No"])
plt.show()
```

### Specifying hue colors

```python

hue_colors = {"Yes": "black", "No": "red"}

sns.scatterplot(x= "total_bill", y= "tip", data=tips, hue= "smoker"
                , palette=hue_colors)
plt.show()
```

### Using HTML hex color codes with hue

```python
import matplotlib.pyplot as plt
import seaborn as sns

hue_colors = {"Yes": "#808080",
                "No": "#00FF00"}

sns.scatterplot(x="total_bill", y="tip", data=tips, hue="smoker",
                palette=hue_colors)

plt.show()
```

### Using hue with count plots

```python
sns.countplot(x="smoker", data=tips, hue="sex")
plt.show()
```
