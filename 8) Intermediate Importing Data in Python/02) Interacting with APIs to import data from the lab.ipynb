## Loading JSONs in Python

```python
# https://www.kaggle.com/yohanan/json-file?select=monthlySalesbyCategoryMultiple.json

import json
with open('./Data/sales.json', 'r') as json_file:
    json_data = json.load(json_file)
```

```python
type(json_data)
```

### Exploring JSONs in Python

```python
for key, value in json_data.items():
    print(key + ':', value)
```
