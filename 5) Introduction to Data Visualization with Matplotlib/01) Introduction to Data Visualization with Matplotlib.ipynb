```python
import pandas as pd
import matplotlib.pyplot as plt

stocks =  pd.read_csv('./Data/all_stocks_5yr.csv')
print(stocks.head())
```

```python
AAL = stocks[stocks.Name == 'AAL']
print(AAL.head())
```

```python
PFE = stocks[stocks.Name == 'PFE']
print(PFE.head())
```

```python
plt.plot(AAL['date'], AAL['open']);
plt.plot(PFE['date'], PFE['open']);
```

### Making Customized Plots

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(AAL['date'], AAL['open'],linestyle="--");
ax.plot(PFE['date'], PFE['open']);


ax.set_xlabel("Time (months)")
ax.set_ylabel("Open Price")
ax.set_title("AAL vs PFE price comparison")
plt.show()
```

```python
# Adding Data AAL

import pandas as pd
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 1)
ax[0].plot(AAL['date'], AAL['open'],linestyle="-", color = 'b');
ax[1].plot(AAL['date'], AAL['close'],linestyle='--', color= 'r');


ax[0].set_xlabel("Time (months)")
ax[0].set_ylabel("Open Price")
ax[1].set_xlabel("Time (months)")
ax[1].set_ylabel("Close Price")

ax[0].set_title("Price comparison");
plt.show()
```

## Small multiples with plt.subplots

```python
# https://stackoverflow.com/questions/37967786/axes-from-plt-subplots-is-a-numpy-ndarray-object-and-has-no-attribute-plot

fig, ax = plt.subplots(2, 1, sharey = True)
ax[0].plot(AAL['date'], AAL['open'],linestyle="-", color = 'b');
ax[0].plot(AAL['date'], AAL['close'],linestyle="-", color = 'r');
plt.show()
```
