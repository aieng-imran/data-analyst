## Plotting time-series with different variables

```python
import pandas as pd
import matplotlib.pyplot as plt

# Watch the video in the link above to find out why we are indexing "date"

stocks = pd.read_csv('./Data/all_stocks_5yr.csv', parse_dates=["date"], index_col = "date")
```

```python
stocks.head()
```

```python
AAL = stocks[stocks.Name == 'AAL']
AAL.shape
```

```python
PFE = stocks[stocks.Name == 'PFE']
PFE.shape
```

## Twin Axes

```python
fig, ax = plt.subplots()

ax.plot(AAL.index, AAL['open'])
ax.plot(PFE.index, PFE['open'])
ax.set_xlabel('Date')
plt.show()
```

```python
# Using Twin Axis

fig, ax = plt.subplots()

ax.plot(AAL.index, AAL['open'])
ax.set_xlabel('Date')
ax.set_ylabel('AAL')

ax2 = ax.twinx()

ax.plot(PFE.index, PFE['open'])
ax2.set_ylabel('PFE')

plt.show()
```

## A function that plots time-series

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

```python
fig, ax = plt.subplots()
plot_timeseries(ax, AAL.index, AAL['open'], 'blue', 'date', 'open-AAL')
ax2 = ax.twinx()
plot_timeseries(ax2, PFE.index, PFE['open'], 'red', 'Time', 'open-PFE')

plt.show()
```

## Annotating timeseries data
