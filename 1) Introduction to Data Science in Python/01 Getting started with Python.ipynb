# How to Load Data

```python

import pandas as pd
import matplotlib.pyplot as plt
# Source of data: https://www.kaggle.com/timoboz/tesla-stock-data-from-2010-to-2020
TSLA_2010_20 = pd.read_csv('./data/TSLA_2010_20.csv')
```

```python
print(TSLA_2010_20)
```

```python
# Plot a graph

x_values = TSLA_2010_20['Date']
y_values = TSLA_2010_20['Close']
plt.plot(x_values, y_values)
plt.xlabel = 'Date'
plt.ylabel = 'Close Price'

# Display the graph
plt.show()
```
