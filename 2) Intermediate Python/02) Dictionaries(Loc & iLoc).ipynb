## Pandas- Part1

```python
dict = {
"country":["Brazil", "Russia", "India", "China", "South Africa"],
"capital":["Brasilia", "Moscow", "New Delhi", "Beijing", "Pretoria"],
"area":[8.516, 17.10, 3.286, 9.597, 1.221],
"population":[200.4, 143.5, 1252, 1357, 52.98] }
```

## Pandas- Part2 ( Index and select data )

```python
# Cleaning Data
fifa = fifa.drop(axis=1, columns = ['ID', 'Photo', 'Club Logo', 'Flag', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking'])
fifa.head()
```

### Column Access

```python
fifa[["Nationality"]]
```

```python
type(fifa[["Nationality"]])
```

```python
fifa[["Nationality", "Age"]]
```

### Row Access

```python
# Remember that the end of the slice is exclusive and that the index starts 
# at zero.

fifa[0:3]
```

```python
# https://sparkbyexamples.com/pandas/pandas-difference-between-loc-vs-iloc-in-dataframe/
```

### Extracting data as "Series"   --- Row Access loc

```python
fifa.loc[1]
```

```python
type(fifa.loc[1])
```

### Extracting data as DataFrame  --- Row Access loc

```python
fifa.loc[[1]]
```

```python
type(fifa.loc[[1]])
```

### Row & Column loc

```python
fifa.loc[0:3, ["Nationality"]]
```

```python
info.loc[["RU", "IN", "CH"], ["country", "capital"]]
```

```python
info.loc[:, ["country", "capital"]]
```

### Row Access iloc

```python
info.loc[["RU", "IN", "CH"]]
```

```python
info.iloc[[1,2,3]]
```

### Row & Column iloc

```python
info.loc[["RU", "IN", "CH"], ["country", "capital"]]
```

```python
info.loc[:, ["country", "capital"]]
```

```python
info.iloc[[1,2,3], [0,1]]
```
