```python
import pandas as pd
import matplotlib.pyplot as plt

fifa = pd.read_csv('./Data/fifa.csv')
```

```python
# Cleaning Data
fifa = fifa.drop(axis=1, columns = ['Unnamed: 0', 'ID', 'Photo', 'Club Logo', 'Flag', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling', 'GKKicking'])
fifa.head()
```

```python
# Clean up value and wage columns

def get_value(value):
    value_num = value.replace('€','')
    if 'M' in value_num:
        value_num = float(value_num.replace('M','')) * 1000000
    elif 'K' in value_num:
        value_num = float(value_num.replace('K','')) * 1000
        
    return float(value_num) # Ensure both columns are in float format

fifa['Value'] = fifa['Value'].apply(lambda x: get_value(x))
fifa['Wage'] = fifa['Wage'].apply(lambda x: get_value(x))

fifa.head()
```

```python
plt.scatter(fifa.Age, fifa.Wage)
plt.xlabel('Age')
plt.ylabel('Wage')
plt.show()
```

### Histogram

```python
plt.hist(fifa.Age, bins=50)
plt.xlabel('Age')
plt.ylabel('Quantity')
plt.title("Age Distribution")
plt.show()
```

```python
# Confirming the above histogram's results
by_age = fifa.Age.value_counts()
type(by_age)
```

### Customization 

```python
# Looking for Average Wage/Age
x = fifa.groupby("Age")["Wage"]
print(x)
```
